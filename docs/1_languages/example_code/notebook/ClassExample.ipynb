```python
import numpy as np
import matplotlib.pyplot as plt

SIZE=100
x = np.arange(SIZE)
```

```python
print(x)
```

```python
y =  x + np.random.random(SIZE) ** 2

y2 = x * np.random.random(SIZE)*10


rect = plt.Rectangle( (20, 200), 20, 100, alpha=0.5, facecolor='g')
plt.gca().add_patch(rect)

rect2 = plt.Rectangle( (60, 200), 20, 100, fill=False, edgecolor='r')
plt.gca().add_patch(rect2)
plt.scatter(x, y)

plt.show()
```

```python
plt.scatter(x, y2)
plt.show()
```

```python
rect = plt.Rectangle( (20, 200), 20, 100, alpha=0.5, facecolor='g')
plt.gca().add_patch(rect)

rect2 = plt.Rectangle( (60, 200), 20, 100, fill=False, edgecolor='r')
plt.gca().add_patch(rect2)

rect3 = plt.Rectangle( (70, 500), 20, 100, fill=False, edgecolor='r', linewidth=3)
plt.gca().add_patch(rect3)


circle = plt.Circle( (10, 100), 10, fc='b')
plt.gca().add_patch(circle)


plt.scatter(x, y)
plt.scatter(x, y2)
plt.show()
```

```python

xAxis = np.arange( 0, 2, 0.1)

yAxis = 12 - xAxis ** 3

plt.subplot( 1, 2, 1 )
plt.plot( xAxis, yAxis, 'r^-', label="x cubed" )
plt.grid( True )

plt.subplot( 1, 2, 2 )
yAxis = 1 + np.cos( 6 * xAxis )
plt.plot( xAxis, yAxis, "go", label="cosine wave" )

plt.title( "A bunch of stuff" )
plt.xlabel( "This is the X Axis" )
plt.ylabel( "This is the Y Axis" )

plt.text( 0.75, np.max( yAxis ), "Sew fancy" )


plt.legend( loc='best' )

plt.show()

```
